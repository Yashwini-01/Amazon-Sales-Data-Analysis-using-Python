```python
import pandas as pd
df = pd.read_csv("amazon.csv")
print(df.shape)
```

```python
print(df.columns)
```

```python
df.head
```

```python
df.info()
df.describe(include="all")
```

```python
df["discounted_price"] = df["discounted_price"].astype(str).str.replace("₹","").str.replace(",","").astype(float)
df["actual_price"] = df["actual_price"].astype(str).str.replace("₹","").str.replace(",","").astype(float)
df["discount_percentage"] = df["discount_percentage"].astype(str).str.replace("%","").astype(float)
df["rating"] = pd.to_numeric(df["rating"], errors="coerce")
df["rating_count"] = df["rating_count"].astype(str).str.replace(",","").astype(float)
```

```python
df.drop(columns=["img_link","product_link"])
```

```python
df = df.dropna(subset=["rating_count"])
```

```python
df["rating"].fillna(df["rating"].mean(), inplace = True)
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df["discount_percentage"], bins=30, kde=True)
plt.title("Distribution of Discount Percentages")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["rating"], bins=20, kde=True)
plt.title("Distribution of Ratings")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="discount_percentage", y="rating", data=df, alpha=0.6)
plt.title("Discount vs Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="actual_price", y="discount_percentage", data=df, alpha=0.6)
plt.title("Price vs Discount Percentage")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatterplot limited to 0 - 20000
plt.figure(figsize=(8,5))
sns.scatterplot(
    x="actual_price",
    y="discount_percentage",
    data=df[df["actual_price"] <= 20000],
    alpha=0.6
)
plt.title("Discount % vs Price (0 - 20,000)")
plt.xlabel("Actual Price (₹)")
plt.ylabel("Discount Percentage")
plt.show()
```

```python
df["price_bin"] = pd.cut(
    df["actual_price"],
    bins=[0, 5000, 10000, 20000, 50000, 100000, df["actual_price"].max()],
    labels=["0-5k", "5k-10k", "10k-20k", "20k-50k", "50k-100k", "100k+"]
)

df.groupby("price_bin")["discount_percentage"].mean()
```

```python
discount_below_20k = df[df["actual_price"] <= 20000]["discount_percentage"].mean()
discount_above_20k = df[df["actual_price"] > 20000]["discount_percentage"].mean()

print("Average discount for <= 20k:", discount_below_20k)
print("Average discount for > 20k :", discount_above_20k)
```

```python
df["main_category"] = df["category"].str.split("|").str[0]
df.groupby("main_category")["discount_percentage"].mean().sort_values(ascending=False)
```
